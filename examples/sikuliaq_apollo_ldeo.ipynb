```python
from datetime import datetime, timezone, timedelta
import matplotlib.pyplot as plt

from crlx import SIKULIAQ
```

```python
skq = SIKULIAQ(verbose = True)
edt = datetime.now(timezone.utc)
bdt = edt - timedelta(days = 3)
```

```python
apollo = skq.get_wet_lab_apollo(bdt, edt)
ldeo = skq.get_wet_lab_ldeo(bdt,edt)
```

```python
fig, ax = plt.subplots(2,1, figsize = (11,8.5), constrained_layout = True)

ax[0].scatter(ldeo.time, ldeo.air_pco2)
ax[0].scatter(apollo.time, apollo.air_pco2)

ax[1].scatter(ldeo.time, ldeo.sea_water_pco2)
ax[1].scatter(apollo.time, apollo.sea_water_pco2)
```
